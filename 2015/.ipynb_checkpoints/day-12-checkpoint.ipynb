```python
import re
```

```python
with open('input/12.txt','r') as f:
    data = f.read()
len(data)
```

```python
# Part 1

pattern = r'-?\d+'
nums = re.findall(pattern,data)
print('Part 1:', sum(list(map(int,nums))))
```

```python
# Part 2

#data = '[1,2,3]'
#data = '[1,{"c":"red","b":2},3,{"a":1,{"bb":1337,"aa":"red"},{"bb":100}},"red",4]'
#data = '''[1,"red",5]'''
#data = '{"d":"red","e":[1,2,3,4],"f":5,"g":}'

sum2 = 0
in_obj = False
in_red = False
in_same_num = False
lvl = 0
lvl_red = 0
hold_sum = 0
for i,c in enumerate(data):
    if c == '{':
        in_obj = True
        lvl += 1
        print('Lvl:', lvl)
    elif c == '}':
        lvl -= 1
        print('Lvl:', lvl)
        if lvl == 0:
            in_obj = False
            in_red = False
        if not in_red and lvl == 0:
            if hold_sum != 0:
                print('}', hold_sum)
                sum2 += hold_sum
                hold_sum = 0
            hold_sum = 0
        if lvl < lvl_red:
            in_red = False
    elif data[i:i+3] == 'red' and in_obj:
        print('Red at lvl:', lvl)
        if hold_sum != 0:
            print('Reset hold_num:', hold_sum)
            hold_sum = 0
        in_red = True
        if lvl < lvl_red:
            lvl_red = lvl
    elif c == ',':
        in_same_num = False
    elif c in '-1234567890' and not in_red and not in_same_num:
        is_num = False
        j = 4
        while not is_num:
            if j == 0:
                break
            try:
                num = int(data[i:i+j])
                hold_sum += num
                is_num = True
                in_same_num = True
            except ValueError:
                j -= 1
        print('Num:', num)
else:
    sum2 += hold_sum
print('Part 2:', sum2)
```
