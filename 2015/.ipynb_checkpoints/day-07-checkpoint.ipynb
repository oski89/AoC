```python
EXAMPLE = 0

if EXAMPLE == 1:
    data = '''123 -> x
    456 -> y
    x AND y -> d
    x OR y -> e
    x LSHIFT 2 -> f
    y RSHIFT 2 -> g
    NOT x -> h
    NOT y -> i'''.split('\n')
else:
    with open('input/07.txt','r') as f:
        data = f.read().split('\n')
#data
```

```python
# Part 1

def parse_instruction(string):
    instr = string.split()
    instr.remove('->')
    return instr

def perform_instruction(instr,vardict):
    outvar = instr[-1]
    if 'AND' in instr:
        invar1 = instr[0]
        invar2 = instr[2]
        if invar1 == '1':
            val = 1 & vardict[invar2]
        elif invar2 == '1':
            val = vardict[invar2] & 1
        else:
            val = vardict[invar1] & vardict[invar2]
    elif 'OR' in instr:
        invar1 = instr[0]
        invar2 = instr[2]
        val = vardict[invar1] | vardict[invar2]
    elif 'LSHIFT' in instr:
        invar = instr[0]
        num = int(instr[2])
        val = vardict[invar] << num
    elif 'RSHIFT' in instr:
        invar = instr[0]
        num = int(instr[2])
        val = vardict[invar] >> num
    elif 'NOT' in instr:
        invar = instr[1]
        val = ~ vardict[invar]
        if val < 0:
            val += 65536
    else:
        try:
            val = vardict[instr[0]]
        except:
            val = int(instr[0])
    vardict[outvar] = val
    return vardict

d = {}
while len(d) != len(data):
    for row in data:
        instr = parse_instruction(row)
        try:
            d = perform_instruction(instr,d)
        except:
            pass

if not EXAMPLE:
    print('Part 1:', d['a'])
else:
    print(d)
```

```python
# Part 2

d2 = {}
while len(d2) != len(data):
    for row in data:
        if row == '1674 -> b':
            row = str(d['a']) + ' -> b'
        instr = parse_instruction(row)
        try:
            d2 = perform_instruction(instr,d2)
        except:
            pass

print('Part 2:', d2['a'])
```
