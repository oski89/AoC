```python
with open('input/03.txt', 'r') as f:
    data = f.read().split()
data
```

```python
# Part 1
nrows = len(data[0])
walked = ''
for n in range(1,len(data)):
    walked += data[1*n][3*n % nrows]
walked
```

```python
print('Part 1:', walked.count('#'))
```

```python
# Part 2
nrows = len(data[0])
directions = [(1,1),(3,1),(5,1),(7,1),(1,2)]
trees_mult = 1
for r,d in directions:
    walked2 = ''
    for n in range(1,len(data)//d):
        walked2 += data[d*n][r*n % nrows]
    trees_mult *= walked2.count('#')
print('Part 2:', trees_mult)
```

```python
# Part 2

nrows = len(data[0])
directions = [(1,1),(3,1),(5,1),(7,1),(1,2)]
trees_mult = 1
for r,d in directions:
    walked2 = ''
    for n in range(1, round(len(data)/d)):
        walked2 += data[d*n][r*n % nrows]
    trees_mult *= walked2.count('#')
    print(walked2.count('#'))
print('Part 2:', trees_mult)
```
