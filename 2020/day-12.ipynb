```python
from itertools import cycle
import numpy as np
```

```python
EX = 0

if not EX:
    with open('input/12.txt','r') as f:
        data = [(c[0],int(c[1:])) for c in f.read().split()]
else:
    data = [(c[0],int(c[1:])) for c in'''F10
            N3
            F7
            R90
            F11'''.split()]
    
#data
```

```python
def navigate(action, coord, ptr):
    
    global ptrs
    
    d = action[0]  # dirction
    s = action[1]  # size
    
    y = coord[0]
    x = coord[1]
    
    if d in 'LR':
        u = (s % 360) // 90
        if d == 'R':
            rot_dir = 4-u
        else:
            rot_dir = u
        for _ in range(rot_dir):
            ptr = next(ptrs)
    
    elif d == 'F':
        d = ptr
    
    if d == 'E':
        x += s
    elif d == 'N':
        y += s
    elif d == 'W':
        x -= s
    elif d == 'S':
        y -= s
        
    coord = y,x
    
    return coord, ptr
```

```python
def waypoint(action, coord, ptr, wpt):
    
    global ptrs
    
    d = action[0]  # dirction
    s = action[1]  # size
    
    y = coord[0]
    x = coord[1]
    
    wy = wpt[0]
    wx = wpt[1]
    
    dy = wy - y
    dx = wx - x
    
    if d in 'LR':
        u = (s % 360) // 90
        rot_mat = np.array([[0,-1],[1,0]])
        np_coord = np.array([x,y]).reshape(2,1)
        np_wpt = np.array([wx,wy]).reshape(2,1)
        
        if d == 'R':
            rot_dir = 4-u
        else:
            rot_dir = u
        for _ in range(rot_dir):
            np_wpt = np.dot(rot_mat,(np_wpt-np_coord)) + np_coord
        wx,wy = map(int,np_wpt)
    
    elif d == 'F':
        y += s*dy
        x += s*dx
        wy = y + dy
        wx = x + dx
    
    elif d == 'E':
        wx += s
    elif d == 'N':
        wy += s
    elif d == 'W':
        wx -= s
    elif d == 'S':
        wy -= s
        
    coord = y,x
    wpt = wy,wx
    
    return coord, ptr, wpt
```

```python
# Part 1

ptrs = cycle('ENWS')
ptr = next(ptrs)
y,x = 0,0
coord = y,x

for action in data:
    coord, ptr = navigate(action, coord, ptr)

man_dist = abs(coord[1]) + abs(coord[0])
    
print('Part 1:', man_dist)
```

```python
# Part 2

ptrs = cycle('ENWS')
ptr = next(ptrs)
y,x = 0,0
wy,wx = 1,10
coord = y,x
wpt = wy,wx

for action in data:
    coord, ptr, wpt = waypoint(action, coord, ptr, wpt)

man_dist = abs(coord[1]) + abs(coord[0])

print('Part 2:', man_dist)
```
