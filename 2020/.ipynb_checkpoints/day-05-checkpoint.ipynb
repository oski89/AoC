```python
with open('input/05.txt','r') as f:
    data = f.read().split()
data
```

```python
# Part 1
max_pass_id = 0
for row in data:
    b,f = 0,128
    r,l = 0,8
    for c in row:
        if c == 'B':
            b = b + (f-b) // 2
        elif c == 'F':
            f = f - (f-b) // 2
        elif c == 'R':
            r = r - (r-l) // 2
        elif c == 'L':
            l = l + (r-l) // 2
    if b*8 + r > max_pass_id:
        max_pass_id = b*8 + r
print('Part 1:', max_pass_id)
```

```python
# Part 2
seat_list = []
for row in data:
    b,f = 0,128
    r,l = 0,8
    for c in row:
        if c == 'B':
            b = b + (f-b) // 2
        elif c == 'F':
            f = f - (f-b) // 2
        elif c == 'R':
            r = r - (r-l) // 2
        elif c == 'L':
            l = l + (r-l) // 2
    seat_list.append(b*8 + r)
print(set(seat_list))
```

```python
my_seat = set(range(40,981)).difference(set(seat_list))
print('Part 2:', my_seat)
```
